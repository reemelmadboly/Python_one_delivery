````python
import pandas as pd
import numpy as np
import re
energy = pd.read_excel('/home/reemelmadboly/ITI_Python_for_Data_Management/Task 1/data/Energy Indicators.xls',
                      header = 1 , skiprows = range(0,16), nrows = 227,  usecols = "C:F")
energy.rename(columns = {'Unnamed: 2': 'Country' , 'Petajoules': 'Energy Supply' , 'Gigajoules':'Energy Supply per Capita' , '%': '% Renewable'} , inplace=True)
#energy.drop(energy.index[[range(0,16)]] ,  inplace=True)
energy.replace('...',np.NaN, inplace =True)
energy['Energy Supply']*=1000000

def removeDigi(str):
    return ''.join([i for i in str if not i.isdigit()])
        
def check_char(str):
    i = str.find('(')
    if(i == -1):
        return str        
    else:
        return str[:i]
    
#print(removeDigi('reem(15'))
energy['Country'] = energy['Country'].apply(removeDigi).apply(check_char)

energy['Country'].replace({"Republic of Korea":"South Korea",
'United States of America':'United States',
'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
'China, Hong Kong Special Administrative Region':'Hong Kong' } , inplace=True)

energy.head()
````
